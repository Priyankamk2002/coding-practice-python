```python
def sortarray(arr):
    arr.sort()

    return arr[0]
```

```python
if __name__ == "__main__":
    arr1 = [2,3,4,5,6,7,8,0,1]
    print(f"The smaleest element in the array is : {sortarray(arr1)}")
```

```python
def smallelement(arr,n):
    mini = arr[0]

    for i in range(1,n):
        if arr[i]<mini:
            mini = arr[i]

    return mini
```

```python
if __name__ == "__main__":
    arr1 = [2,4,5,6,7,8,9,10,0,1]
    n = len(arr1)

    mini = smallelement(arr1,n)
    print(f"The smallest element in the array is : {mini}")
```

```python
def palindrome(n):
    revnum = 0
    dup = n

    while n>0:
        ld = n%10
        revnum = (revnum*10)+ld
        n = n//10

    return dup==revnum

number = 4554
if palindrome(number):
    print(f"{number} is a palindrome")
else:
    print(f"{number} is not a palindrome")
```

```python
a,b = map(int,input().split())
start = min(a,b)
end = max(a,b)

sum_cubes = 0

i = start
while i<=end:
    sum_cubes = sum_cubes + (i**3)
    i+=1

print(sum_cubes)
```

```python
start,end = map(int,input().split())
cube_sum = sum(i**3 for i in range(start,end+1))
print(cube_sum)
```
